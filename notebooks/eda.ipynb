```python
# number of images each category has

classes_dict = {}

for sampling in os.listdir(data):
    sampling = os.path.join(data, sampling)
    for image_class in os.listdir(sampling):
        image_class = os.path.join(sampling, image_class)
        if os.path.basename(image_class) not in classes_dict:
            classes_dict[os.path.basename(image_class)] = [] 
        classes_dict[os.path.basename(image_class)].append(len(os.listdir(image_class)))

print(f'{'Emotion': <10} {'Train': >10} {'Test': >10} {'Total': >10}')
print('-' * 43)
for category in classes_dict:
    print(f'{category: <10} {classes_dict[category][1]: >10} {classes_dict[category][0]: >10} {sum(classes_dict[category]): >10}')
```

```python
# Distribution of images per emotions
print(f'{'Emotion': <10} {'Train': >10} {'Test': >10} {'Total': >10}')

sum_train = 0
sum_test = 0
sum_total = 0
for numbers in classes_dict.values():
    sum_train += numbers[1]
    sum_test += numbers[0]
    sum_total += sum(numbers)

print('-' * 43)
for key, values in classes_dict.items():
    train_percentage = (values[1] / sum_train) * 100
    test_percentage = (values[0] / sum_test) * 100
    sum_percentage = (sum(values) / sum_total) * 100

    print(f'{key: <10} {train_percentage :>10.1f} {test_percentage :>10.1f} {sum_percentage :>10.1f}')
```

```python
# Displaying images for each category
import random

train = os.path.join(data, 'train')

emotions = [item for item in os.listdir(train)]
for emotion in emotions:
    print(emotion.title())
    images_dir = os.path.join(train, emotion)
    images_list = os.listdir(images_dir)

    indexes = list(range(len(images_list)))
    random.shuffle(indexes)
    indexes = indexes[:3]

    for i in indexes:
        img = mpimg.imread(os.path.join(images_dir, images_list[i]))
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        plt.show()
```
